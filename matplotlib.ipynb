```python
import matplotlib.pyplot as plt # 导入 matplotlib 软件包中的 pyplot 模块并简写为 plt
import numpy as np # 导入 numpy 软件包并简写为 np
x = np.linspace(-3,3,50) # 定义一个 -3 到 3 的等差数列，项数为 50
y = 2*x+1
plt.figure(num=1,figsize=(8,5)) # 在这个 figure1 中显示图像，设置图片长宽分别为 8 和 5 
plt.plot(x,y) # 画出这条直线
ax = plt.gca() # 获得当前坐标系
ax.spines['right'].set_color('none') # 去掉最右边的坐标轴
ax.spines['top'].set_color('none') # 去掉最上面的坐标轴
ax.xaxis.set_ticks_position('bottom') # 将最下面的坐标轴轴设置为 x 轴
ax.spines['bottom'].set_position(('data',0)) # 设置将坐标原点位置的 y 值为 0
ax.yaxis.set_ticks_position('left') # 将最左边的坐标轴设置为 y 轴
ax.spines['left'].set_position(('data',0)) # 设置坐标原点位置的  x 值为 0 
x0 = 1 
y0 = 2*x0+1
plt.scatter(x0,y0,s=50,color='r') # 显示这个点并且设置这个点的大小为 50（s=50）颜色为红色（color='r'）
plt.plot([x0,x0],[0,y0],'k--',lw=2.5) # 先给出线段两个端点的 x y 的值，画黑色的虚线，线宽为 2.5
# 2x+1=%s 是我们需要添加的注释内容，前后两个 $ 是为了美化字符
# 设置 xy 的值为 (x0,y0) 就是我们要标注的点的坐标
# xycoords='data' 表示以我们设置的 xy 的值为基准位置，这里的 data 就代表 xy
# xytext=(+30,-30) textcoords='offset point' 表示在基准位置右边 30 下面 30 的位置写入标注内容
# fontsize=16 设置字体的大小是 16
# arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad.2')) 就是用来画箭头的

# 在 x=-3.7 y=3 的位置上用大小 16 绿色的字打出标注，可以打出一些数学符号 
plt.text(-3.7,3,r'$This\ is\ the\ some\ text.\mu\ \sigma_i\ \alpha_t$',
         fontdict={'size':16,'color':'g'})
plt.show()
```

```python
import matplotlib.pyplot as plt # 导入 matplotlib 软件包中的 pyplot 模块并简写为 plt
import numpy as np # 导入 numpy 软件包并简写为 np
x = np.linspace(-3,3,50) # 定义一个 -3 到 3 的等差数列，项数为 50
y = 0.1*x
plt.figure() # 在这个 figure 中显示图像
plt.plot(x,y,linewidth=10,alpha=0.7) # 画出这条直线,设置线宽为 10
# alpha 是设置透明度的一个参数设置透明度这个括号里的值应该在 0 到 1 之间，越接近 0 约透明
plt.ylim(-2,2) # 把 y 轴限制在 -2 到 2之间
ax = plt.gca() # 获得当前坐标系
ax.spines['right'].set_color('none') # 去掉最右边的坐标轴
ax.spines['top'].set_color('none') # 去掉最上面的坐标轴
ax.xaxis.set_ticks_position('bottom') # 将最下面的坐标轴轴设置为 x 轴
ax.spines['bottom'].set_position(('data',0)) # 设置将坐标原点位置的 y 值为 0
ax.yaxis.set_ticks_position('left') # 将最左边的坐标轴设置为 y 轴
ax.spines['left'].set_position(('data',0)) # 设置坐标原点位置的  x 值为 0 
plt.show()  
```

```python
import matplotlib.pyplot as plt # 导入 matplotlib 软件包中的 pyplot 模块并简写为 plt
import numpy as np # 导入 numpy 软件包并简写为 np
x = np.linspace(-3,3,50) # 定义一个 -3 到 3 的等差数列，项数为 50
y = 0.1*x
plt.figure() # 在这个 figure 中显示图像
plt.plot(x,y,linewidth=10) # 画出这条直线,设置线宽为 10
plt.ylim(-2,2) # 把 y 轴限制在 -2 到 2之间
ax = plt.gca() # 获得当前坐标系
ax.spines['right'].set_color('none') # 去掉最右边的坐标轴
ax.spines['top'].set_color('none') # 去掉最上面的坐标轴
ax.xaxis.set_ticks_position('bottom') # 将最下面的坐标轴轴设置为 x 轴
ax.spines['bottom'].set_position(('data',0)) # 设置将坐标原点位置的 y 值为 0
ax.yaxis.set_ticks_position('left') # 将最左边的坐标轴设置为 y 轴
ax.spines['left'].set_position(('data',0)) # 设置坐标原点位置的  x 值为 0 

# 采用循环的方式对坐标轴每个数字的大小透明度进行设置
for label in ax.get_xticklabels()+ax.get_yticklabels():
    label.set_fontsize(22) # 设置坐标轴数字的大小
    label.set_alpha(0.2)# 设置坐标轴数字透明度，这个括号里的值应该在 0 到 1 之间，越接近 0 约透明越接近 1 越不透明
plt.show()
```

```python
import matplotlib.pyplot as plt # 导入 matplotlib 软件包中的 pyplot 模块并简写为 plt
import numpy as np # 导入 numpy 软件包并简写为 np
n = 1024 
X = np.random.normal(0,1,n) # 随机生成 n 个均值为 0 方差为 1 的正态分布值，并赋给 X
Y = np.random.normal(0,1,n) #  随机生成 n 个均值为 0 方差为 1 的正态分布值，并赋给 Y
T = np.arctan2(Y,X) # 设置颜色
plt.scatter(X,Y,75,c=T,alpha=0.5) # 画散点图，设置点的大小为 75，透明度为 0.5
plt.xlim((-1.5,1.5)) # 将 x 轴限制在 -1.5 到 1.5 的范围内
plt.ylim((-1.5,1.5)) # 将 y 轴限制在 -1.5 到 1.5 的范围内
plt.xticks(()) # 隐藏 x 轴
plt.yticks(()) # 隐藏 y 轴
plt.show() # 显示图片
```
